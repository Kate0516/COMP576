```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
x = np.arange(-2,2,0.00001)
y = np.sqrt(np.power(2,2) - np.power(x,2))
plt.plot(x,y)
plt.plot(x,-y)
plt.show()
```
